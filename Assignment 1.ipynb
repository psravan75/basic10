3. Compare and contrast mutable and immutable objects in Python with examples 

->>. "Mutable vs. Immutable Objects in Python"

- *Mutable Objects*:
  - *Can be changed* after creation.
  - *Examples*: Lists, dictionaries, sets.
  - *Example*:
    python
    my_list = [1, 2, 3]
    my_list.append(4)  # my_list is now [1, 2, 3, 4]
    

- *Immutable Objects*:
  - *Cannot be changed* after creation; any modification creates a new object.
  - *Examples*: Strings, tuples, integers.
  - *Example*:
    python
    my_string = "Hello"
    new_string = my_string + " World"  # my_string is still "Hello"
    

### *Key Differences*:
- *Modification*: Mutable objects change in place; immutable objects require a new object.
- *Hashability*: Immutable objects can be used as dictionary keys; mutable objects cannot.

4. Discuss the different types of operators in Python and provide examples of how they are used.                       *Types of Operators in Python*:

1. *Arithmetic Operators*: Perform mathematical operations.
   - *Examples*: + (addition), - (subtraction), * (multiplication), / (division), % (modulus)
   - *Example*:
     python
     a = 10
     b = 5
     result = a + b  # result is 15
     

2. *Comparison Operators*: Compare values and return a boolean result.
   - *Examples*: == (equal), != (not equal), > (greater than), < (less than), >= (greater than or equal), <= (less than or equal)
   - *Example*:
     python
     x = 10
     y = 5
     is_equal = x == y  # is_equal is False
     

3. *Logical Operators*: Combine boolean values.
   - *Examples*: and, or, not
   - *Example*:
     python
     a = True
     b = False
     result = a and not b  # result is True
     

4. *Assignment Operators*: Assign values to variables and modify their values.
   - *Examples*: = (assignment), += (add and assign), -= (subtract and assign)
   - *Example*:
     python
     x = 10
     x += 5  # x is now 15
     

5. *Bitwise Operators*: Perform operations on binary representations.
   - *Examples*: & (and), | (or), ^ (xor), ~ (not), << (left shift), >> (right shift)
   - *Example*:
     python
     a = 5  # 0101 in binary
     b = 3  # 0011 in binary
     result = a & b  # result is 1 (0001 in binary)
     

6. *Membership Operators*: Check for membership in sequences.
   - *Examples*: in, not in
   - *Example*:
     python
     my_list = [1, 2, 3]
     exists = 2 in my_list  # exists is True
     

7. *Identity Operators*: Compare the memory locations of objects.
   - *Examples*: is, is not
   - *Example*:
     python
     x = [1, 2]
     y = x
     same = x is y  # same is True
     

These operators help in performing various operations and comparisons in Python programs.

5. Explain the concept of type casting in Python with examples

->> *Type casting* in Python involves converting one data type to another, either automatically or manually. 

### *Implicit Casting*:
- *Automatic Conversion*: Python does this to prevent data loss.
- *Example*: Adding an int and a float results in a float.

### *Explicit Casting*:
- *Manual Conversion*: Use functions to convert types.
- *Examples*:
  - float(5) turns 5 into 5.0.
  - int(5.7) turns 5.7 into 5.
  - str(10) turns 10 into "10".

Type casting ensures that different data types are compatible for operations and functions.

6.  How do conditional statements work in Python? Illustrate with examples

->> Conditional statements in Python let you control the flow of your program by making decisions based on conditions. Here’s how they work:

1. **if Statement**:
   - *Purpose*: Runs a block of code if a condition is true.
   - *Example*:
     python
     score = 85
     if score >= 60:
         print("You passed!")
     
     - *Explanation*: If score is 60 or more, it prints "You passed!"

2. **else Statement**:
   - *Purpose*: Runs if the if condition is false.
   - *Example*:
     python
     score = 45
     if score >= 60:
         print("You passed!")
     else:
         print("You failed.")
     
     - *Explanation*: If score is less than 60, it prints "You failed."

3. **elif (Else If) Statement**:
   - *Purpose*: Checks multiple conditions. Runs the block of code for the first true condition.
   - *Example*:
     python
     temperature = 10
     if temperature < 0:
         print("It's freezing!")
     elif temperature < 20:
         print("It's cold.")
     else:
         print("It's warm.")
     
     - *Explanation*: It prints "It's cold." because the temperature is between 0 and 20.

In summary, if checks a condition, else handles the case when the condition is false, and elif allows checking additional conditions.

7.  Describe the different types of loops in Python and their use cases with examples

->> In Python, loops allow you to repeat code:

### **1. for Loop**:
- *Purpose*: Iterates over a sequence (like a list or range).
- *Example*:
  python
  for i in range(3):
      print(i)
  
  - *Explanation*: Prints numbers from 0 to 2.

### **2. while Loop**:
- *Purpose*: Repeats code as long as a condition is true.
- *Example*:
  python
  count = 0
  while count < 3:
      print(count)
      count += 1
  
  - *Explanation*: Prints numbers from 0 to 2.

### *Control Statements*:
- **break**: Exits the loop early.
- **continue**: Skips to the next iteration.
- **else**: Runs after the loop finishes if not interrupted by break.
